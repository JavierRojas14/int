```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import json
```

```python
df = pd.read_excel('asociacion_bodega_sigfe.xlsx', header = 3)
df['Código'] = df['Código'].astype(str).str.strip().str.upper()
df['Descripción'] = df['Descripción'].astype(str).str.strip().str.upper()
```

## Analizar códigos/descripciones redundantes

```python
codigo_descripcion_agrupados = df.groupby(by = ['Código', 'Descripción'])['Items'].sum()
codigos_repetidos = codigo_descripcion_agrupados.index.get_level_values(0).duplicated(keep = False)
codigos_redundantes = codigo_descripcion_agrupados[codigos_repetidos]
codigos_redundantes.to_excel('codigos_redundates.xlsx')
print(f'Hay {len(codigos_redundantes.index.get_level_values(0))} códigos redundantes')
```

```python
descripcion_codigo_agrupados = df.groupby(by = ['Descripción', 'Código'])['Items'].sum()
descripciones_repetidas = descripcion_codigo_agrupados.index.get_level_values(0).duplicated(keep = False)
descripciones_redundantes = descripcion_codigo_agrupados[descripciones_repetidas]
descripciones_redundantes.to_excel('descripciones_redundantes.xlsx')
print(f'Hay {len(descripciones_redundantes.index.get_level_values(0))} descripciones redundates')
```
