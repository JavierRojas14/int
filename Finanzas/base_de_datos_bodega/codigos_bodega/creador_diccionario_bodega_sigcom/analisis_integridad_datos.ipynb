```python
import pandas as pd
```

```python
df = pd.read_excel('asociacion_bodega_sigfe.xlsx', header = 3)
df['Código'] = df['Código'].astype(str)
df['Descripción'] = df['Descripción'].astype(str)
```

```python
print(f'Hay {len(df["Código"].unique())} códigos únicos')
print(f'Hay {len(df["Descripción"].unique())} descripciones únicas')
```

```python
codigo_descripcion_agrupados = df.groupby(by = ['Familia', 'Código', 'Descripción'])['Items'].sum()
codigos_repetidos = codigo_descripcion_agrupados.index.get_level_values(1).duplicated(keep = False)
codigos_redundantes = codigo_descripcion_agrupados[codigos_repetidos]
codigos_redundantes.to_excel('codigos_redundates.xlsx')
print(f'Hay {len(codigos_redundantes.index.get_level_values(1))} códigos redundantes')
```

```python
descripcion_codigo_agrupados = df.groupby(by = ['Familia', 'Descripción', 'Código'])['Items'].sum()
descripciones_repetidas = descripcion_codigo_agrupados.index.get_level_values(1).duplicated(keep = False)
descripciones_redundantes = descripcion_codigo_agrupados[descripciones_repetidas]
descripciones_redundantes.to_excel('descripciones_redundantes.xlsx')
print(f'Hay {len(descripciones_redundantes.index.get_level_values(1))} descripciones redundates')
```
