```python
import pandas as pd
import json
```

```python
def obtener_las_que_tienen_ref_a_factura_electronica(lista_referencias):
    if lista_referencias:
        for factura in lista_referencias:
            if factura['Tipo'] == '33':
                return True
        return False
    
    else:
        return False
```

```python
def obtener_la_referencia(lista_referencias):
    if lista_referencias:
        for factura in lista_referencias:
            if factura['Tipo'] == '33':
                return factura['Folio']
```

```python
df = pd.read_excel('input\\reporte_97131163.xls')
```

```python
df_2 = df.copy()
df_2['tipo'] = df_2['tipo'].astype(str)
df_2['referencias'] = df_2['referencias'].apply( \
                    lambda x: json.loads(x, strict = False) if type(x) == str else None)
df_sin_nc_ni_gd = df_2.query('not(tipo == "61" or tipo == "52")')
```

```python
mask_tienen_referencia_fe = df_sin_nc_ni_gd['referencias'].apply(obtener_las_que_tienen_ref_a_factura_electronica)
df_posibles_refacturaciones = df_sin_nc_ni_gd[mask_tienen_referencia_fe]
df_posibles_refacturaciones.head(5)
```

```python
df_posibles_refacturaciones['referencia_refacturacion'] = df_posibles_refacturaciones['referencias'].apply(obtener_la_referencia)
df_posibles_refacturaciones['llave_id_ref'] = df_posibles_refacturaciones['emisor'] + df_posibles_refacturaciones['referencia_refacturacion']
```
