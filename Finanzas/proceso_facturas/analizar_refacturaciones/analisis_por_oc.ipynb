```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_excel('input\\PLANILLA DE CONTROL AL 2022-10-13.xlsx')
```

```python
df_sin_SCI_TURBO = df.query('`Fecha Recepción SCI`.isna() and `Folio_interno TURBO`.isna()')
mask_oc_validas = ~df_sin_SCI_TURBO['folio_oc ACEPTA'].isin(['ANULA', '.', np.nan])
df_sin_SCI_TURBO_oc_validas = df_sin_SCI_TURBO[mask_oc_validas]
```

```python
df_a_preguntar = pd.DataFrame()

for oc in df_sin_SCI_TURBO_oc_validas['folio_oc ACEPTA'].unique():
    df_oc = df_sin_SCI_TURBO_oc_validas.query('`folio_oc ACEPTA` == @oc')
    if (df_oc.shape[0] > 1) and df_oc['Fecha Reclamo SII'].any():
        df_facturas_a_preguntar = df_oc.query('REFERENCIAS.isna()')
        df_a_preguntar = pd.concat([df_a_preguntar, df_facturas_a_preguntar])
```
