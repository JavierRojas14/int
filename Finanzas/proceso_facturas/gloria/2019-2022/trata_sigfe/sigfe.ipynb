```python
import pandas as pd
```

```python
df = pd.read_excel('Sigfe_2019-2022.xlsx')
```

```python
mask_debe_sin_nc = (df['Debe'] != 0) & (df['Tipo Documento'] != 'Nota de Credito')
mask_debe_con_nc = (df['Debe'] != 0) & (df['Tipo Documento'] == 'Nota de Credito')

mask_haber_sin_nc = (df['Haber'] != 0) & (df['Tipo Documento'] != 'Nota de Credito')
mask_haber_con_nc = (df['Haber'] != 0) & (df['Tipo Documento'] == 'Nota de Credito')
```

```python
df.loc[mask_debe_sin_nc, 'FOLIO DEBE SIN NC'] = df['Folio'][mask_debe_sin_nc]
df.loc[mask_debe_con_nc, 'FOLIO DEBE CON NC'] = df['Folio'][mask_debe_con_nc]
df.loc[mask_haber_sin_nc, 'FOLIO HABER SIN NC'] = df['Folio'][mask_haber_sin_nc]
df.loc[mask_haber_con_nc, 'FOLIO HABER CON NC'] = df['Folio'][mask_haber_con_nc]
```
