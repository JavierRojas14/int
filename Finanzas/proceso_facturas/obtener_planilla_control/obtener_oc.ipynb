```python
import pandas as pd
from time import sleep
import requests
import json
import numpy

TICKET_MERCADO_PUBLICO = '7CA7E3D8-361B-415F-84EB-88C0B89838B5'
```

```python
def funcion_obtener_requests_mercado_publico(orden_de_compra):
    '''
    Con esta funci√≥n se obtie el detalle de las ordenes de compra que necesitan ser
    analizadas!.'''
    print(f'Pidiendo la orden de compra: {orden_de_compra}')
    orden_de_compra = str(orden_de_compra).strip()
    url_request = f"https://api.mercadopublico.cl/servicios/v1/publico/ordenesdecompra." \
                    f"json?codigo={orden_de_compra}&ticket={TICKET_MERCADO_PUBLICO}"

    try:
        response = requests.get(url_request, timeout = 20)
        detalle_oc = response.json()

        sleep(2.0)

    except Exception as excepcion:
        a_printear = f'{type(excepcion)}: {excepcion}'
        print(a_printear)

        detalle_oc = a_printear

    return detalle_oc
```

```python
df = pd.read_excel('PLANILLA DE CONTROL AL 2022-11-03.xlsx')
```

```python
mask_oc_a_pedir = df['folio_oc ACEPTA'].notna() & df['detalle_oc'].isna()
df_oc_a_pedir = df[mask_oc_a_pedir]
```

```python
for oc in df_oc_a_pedir['folio_oc ACEPTA'].unique():
    mask_oc = df['folio_oc ACEPTA'] == oc
    detalle_oc = funcion_obtener_requests_mercado_publico(oc)
    df.loc[mask_oc, 'detalle_oc'] = json.dumps(detalle_oc, ensure_ascii=False, indent=2)
```

```python
df.to_excel('PLANILLA DE CONTROL AL 2022-11-03.xlsx', index=False)
```
